Task6: Create a pyspark dataframe df1 reading the data from the above file using pyspark magic command %python

```python
df1 = spark.read.csv("/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv", header=True)
#print(df1.is_cached)
#df2=df1.repartition(4)
#display(df2)
display(df1)
```

```python
spark.sql("""
ALTER TABLE mobile_os_usage
ADD COLUMNS (
    Date STRING,
    Mobile_Operating_System STRING,
    Percent_of_Usage STRING
)
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS mobile_os_usage1 (
    Date STRING,
    Mobile_Operating_System STRING,
    Percent_of_Usage STRING
)
""")
```

```python
spark.sql("""
INSERT INTO mobile_os_usage1 (Date, Mobile_Operating_System, Percent_of_Usage) VALUES
('2025-01-01', 'Android', '72.5'),
('2025-01-01', 'iOS', '26.8'),
('2025-01-01', 'Other', '0.7')
""")
```

```python
df3 = spark.sql("SELECT * FROM default.mobile_os_usage1")
display(df3)
```

Task7: Write the above dataframe df1 data into a databricks table called 'default.mobile_os_usage' using pyspark magic command %python

```python
df2 = df1.withColumnRenamed(
    "Mobile Operating System", "Mobile_Operating_System"
).withColumnRenamed(
    "Percent of Usage", "Percent_of_Usage"
)
display(df2)
df2.write.mode("overwrite").saveAsTable("mobile_os_usage")
```

Task8: Write sql query to display the data loaded into the table 'default.mobile_os_usage' using the pyspark magic command %python

```python
df = spark.sql("SELECT * FROM default.mobile_os_usage")
display(df)
```
